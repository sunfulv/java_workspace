#### 2.守护线程(daemon thread)

在Java中有两类线程：User Thread(用户线程)、Daemon Thread(守护线程) 

用个比较通俗的比如，任何一个守护线程都是整个JVM中所有非守护线程的保姆：

**只要当前JVM实例中尚存在任何一个非守护线程没有结束，守护线程就全部工作；只有当最后一个非守护线程结束时，守护线程随着JVM一同结束工作。**  
Daemon的作用是为其他线程的运行提供便利服务，守护线程最典型的应用就是 GC (垃圾回收器)，它就是一个很称职的守护者。

User和Daemon两者几乎没有区别，唯一的不同之处就在于虚拟机的离开：如果 User Thread已经全部退出运行了，只剩下Daemon Thread存在了，虚拟机也就退出了。 因为没有了被守护者，Daemon也就没有工作可做了，也就没有继续运行程序的必要了。

```java
//在这个例子中，设置一个守护线程t,但是守护线程要延时100ms在执行，在这100ms中，主线程已经执行完退出了，
//这时由于没有其他的用户线程了，虚拟机直接退出，所以守护线程还没执行完成就终止了 

class TestRunnable implements Runnable{
    public void run() {
        try{
            Thread.sleep(100);//先休眠1ms,在输出
            System.out.println("当前执行的线程为:"+Thread.currentThread().getName());
        }catch(InterruptedException e){
            e.printStackTrace();
        }
    }
}

TestRunnable r = new TestRunnable();
Thread t = new Thread(r);
t.setDaemon(true);//将线程t设置为守护线程
t.start();
System.out.println("主线程的名字为:"+Thread.currentThread().getName());
```

##### 1.重入锁 ReentrantLock
重入锁的意思是这个锁能够支持一个线程对资源重复加锁。（重入锁还支持获取锁时的公平和非公平性选择）  
也就是在调用**锁的lock()方法**时，对于已经获取到该锁的线程 ，能够再次调用lock()方法而不被阻塞  
**也就是说线程可以重复获得已经持有的锁**，锁保持一个持有计数(hold count)来跟踪方法的嵌套调用

```java
import java.util.concurrent.locks.ReentrantLock;
class Count {
    int count;
    ReentrantLock myLock = new ReentrantLock();
    public Count(int i){
        count = i;
    }
    public int countAddOne(){//写入数据,并返回新的值
        //myLock.lock();
        try{
                count++;
                return count;
        }finally{
          //      myLock.unlock();
        }
    }
}

class Main{
    public static void main(){
        Count c = new Count(10);
        Runnable r = ()->{  try{
                            int i = 0;
                          while(i++<2){
                              int currentCount = c.countAddOne();

                              System.out.println("当前线程为:"+Thread.currentThread().getName()+"线程状态为:"+Thread.currentThread().getState()+"当前计数器c的值为:"+currentCount);
                              Thread.sleep(5);
                            }}catch(InterruptedException e){
                                e.printStackTrace();
                            }
                        };

        for(int i=0;i<10;i++){
            Thread t = new Thread(r);
            t.start();
        }
    }
}
Main.main();
Thread.sleep(1000);
```

#### 14.5.4 条件对象（条件变量）
线程除了会共同竞争某一个资源的使用权(使用锁机制来保证互斥资源同一时刻只有一个线程使用)  
有时还需要**相互合作**来完成特定的任务(线程按照一定的顺序执行)，**某一个线程在另一个线程执行完成后，才能继续执行**

java中，使用等待/通知机制来实现这个功能  


**java的Condition对象**  
Condition用于管理已经获得了一个锁但是却不能做有用工作的线程。由Lock对象的newCondition()方法一个与当前Lock对象相关联的条件对象，共同管理该锁相关的对象和线程。

当线程已经获得锁，但是因为条件限制，无法继续工作的时候，调用Condition.await()方法，此线程进入该条件对象的等待集。当锁可用时，该线程不能马上解除阻塞。相反，它处于阻塞状态，直到另个线程调用同一条件上的signalAll方法时为止，这个方法可以激活所有等待的线程。Condition还有一个Signal方法，该方法是随机激活一个线程。
当一个线程调用awiat()方法时，它无法重新激活自身。 只能寄希望给予其他的线程。如果没有其他线程激活，那他就永远无法在运行了，这就是死锁现象。
SignalAll不会立即激活一个等待线程，它仅仅解除等待线程的阻塞，以便这些线程可以在当前线程退出同步方法之后，通过竞争实现对对象的访问

**Condition对象的await()方法和singal()/signalAll()方法**

Lock锁对象可以创建多个Condition条件对象。每个条件对象，才能调用signal()和await()方法  
当线程如果因为某个条件不满足(使用while循环来不断的检查是否满足条件).条件对象调用await()方法，然后线程进入这个条件的等待队列，然后释放锁。  当锁可用的时候，线程不能马上解除阻塞，需要另一个线程调用同一个条件对象上的signalAll()方法时，才能激活该线程，这是线程又变得可运行态，重新试图获得锁。获得锁止呕可以从await()返回，然后从组册的位置继续运行。

```java
//实现一个有界队列，如果队列为空，线程获取元素操作会导致线程堵塞，直到队列不空。
//如果队列满，线程插入元素会导致线程阻塞，知道队列中出现空位。
import java.util.concurrent.locks.ReentrantLock;
import java.util.concurrent.locks.Condition;
public class BoundedQueue<T>{
    Object[] queue;
    ReentrantLock myLock = new ReentrantLock();
    Condition emptyCondition = myLock.newCondition();
    Condition fullCondition = myLock.newCondition();
    int size;
    int front;
    int rear;
    public BoundedQueue(int length) {
        queue = new Object[length+1];//预留出一个空位置
        size = 0;
    }
    
    public boolean isEmpty(){
        if(size == 0) return true;
        return false;
    }
    public void add(T ele){
        myLock.lock();//先获取锁
        try{
                while(size == queue.length-1)//循环检测队列是否满
                    emptyCondition.await();//等待队列出现空闲
                queue[rear] = ele;
                size++;
                rear = (rear+1)%queue.length;
                System.out.println("当前线程id为:"+Thread.currentThread().getId()+" 当前插入的元素为:"+ele+" 当前队列元素:"+Arrays.toString(queue));
                
                fullCondition.signalAll();//队列中出现了元素，唤醒那些因为队列没有元素而阻塞的线程
            }catch(InterruptedException e){
                e.printStackTrace();
            }finally{
                myLock.unlock();//操作完成释放锁
            }
    }
    @SuppressWarnings("unchecked")
    public T delete(){
        myLock.lock();
        T oldValue = null;
        try{
                while(size == 0)//循环检测队列是否为空,这是为了防止过早或者意外的通知
                    fullCondition.await();
                oldValue = (T)queue[front];
                queue[front] = null;
                front = (front+1)%queue.length;
                size--;
                System.out.println("当前线程id为:"+Thread.currentThread().getId()+" 删除的元素为:"+oldValue+"当前队列元素为:"+Arrays.toString(queue));
                emptyCondition.signalAll();
            }catch(InterruptedException e){
                e.printStackTrace();
            }finally{
                myLock.unlock();
            }
        return oldValue;
    }
}

BoundedQueue<Integer> myQueue = new BoundedQueue(5);
Runnable r1 =()->{
    for(int i=0;i<10;i++){
        myQueue.add(i*i);
        try{
            Thread.sleep(10);
        }catch(InterruptedException e){
            e.printStackTrace();
        }
    }
};
Runnable r2 = ()->{
    for(int i = 0;i<10;i++){
        myQueue.delete();
        try{
            Thread.sleep(10);
        }catch(InterruptedException e){
            e.printStackTrace();
        }
    }
};

Thread t1 = new Thread(r1);
Thread t2 = new Thread(r2);
Thread t3 = new Thread(r2);
System.out.printf("当前创建的子线程:%d.%d,%d\n",t1.getId(),t2.getId(),t3.getId());
t3.start();
t2.start();
t1.start();

    
Thread.sleep(1000);
```

#### 14.5.5 synchronized关键字

Lock和Condition接口为程序提供了高度的锁定控制。  
此外java内部提供了一种内部机制来实现同步。  
Java对象中每个对象都有一个内部锁，如果一个方法使用synchronized关键字声明，那么对象的锁将保护整个方法。**要调用这个方法，线程必须获得内部锁。**

```java
class Counter{
    int count;
    public Counter(int i){
        count = i;
    }
    public synchronized void addOne(){
            count++;
            System.out.println("当前线程id为:"+Thread.currentThread().getId()+" 执行加一操作 "+"当前计数器的值为:"+count);
            notify(); //计数器大于0，唤醒阻塞的队列
            //Thread.sleep(1);

    }
    public synchronized void minusOne(){
        try{
            while(count<=0)
                wait();//如果计数器小于0，等待
            count--;
            System.out.println("当前线程id为:"+Thread.currentThread().getId()+" 执行减一操作"+" 当前计数器的值为:"+count);
            //Thread.sleep(1);
        }catch(InterruptedException e){
            e.printStackTrace();
        }
    } 
}

Counter c = new Counter(0);
Runnable r1 = ()->{
    for(int i=0;i<100;i++){
        c.addOne();
    }
};
Runnable r2 = ()->{
    for(int j=0;j<100;j++){
        c.minusOne();
    }
};
Thread t1 = new Thread(r1);
Thread t2 = new Thread(r2);
t2.start();
t1.start();
Thread.sleep(1000);
```

#### 14.5.6 volatile 关键字

　　一旦一个共享变量（类的成员变量、类的静态成员变量）被volatile修饰之后，那么就具备了两层语义：

　　1）保证了不同线程对这个变量进行操作时的可见性，即一个线程修改了某个变量的值，这新值对其他线程来说是立即可见的。   
　　2）禁止进行指令重排序。
  
  
**volatile关键字是与Java的内存模型有关的，因此在讲述volatile关键之前，我们先来了解一下与内存模型相关的概念和知识**

##### 1.内存模型的相关概念
当程序在运行过程中，会将运算需要的数据**从主存复制一份到CPU的高速缓存**当中，那么CPU进行计算时就可以直接从它的高速缓存读取数据和向其中写入数据，当运算结束之后，再将高速缓存中的数据刷新到主存当中。

但是对于多核系统，每个CPU都有自己的高速缓存，如果变量在多个高速缓存中都有缓存(多线程编程中)，**会导致缓存不一致问题**  
为了解决缓存不一致问题，一种可行的办法是定义**缓存一致性协议。**  


缓存一致性协议。最出名的就是Intel 的MESI协议，MESI协议保证了每个缓存中使用的共享变量的副本是一致的。它核心的思想是：**当CPU写数据时，如果发现操作的变量是共享变量，即在其他CPU中也存在该变量的副本，会发出信号通知其他CPU将该变量的缓存行置为无效状态，因此当其他CPU需要读取这个变量时，发现自己缓存中缓存该变量的缓存行是无效的，那么它就会从内存重新读取**

##### 2.并发编程的三个问题

>原子性
>可见性
>有序性

1. 原子性
即一个操作或者多个操作 要么全部执行并且执行的过程不会被任何因素打断，要么就都不执行。  
2. 可见性
即一个操作或者多个操作 要么全部执行并且执行的过程不会被任何因素打断，要么就都不执行。
3. 有序性
有序性：即程序执行的顺序按照代码的先后顺序执行。  
**但是有时编译器会在不改变最终运算结果的情况下改变指令的执行顺序(也就是指令重排序)**  
但是如果是多线程程序，改变指令的执行顺序可能会导致最终结果发生改变，从而得不到正确结果。    
**指令重排序不会影响单个线程的执行，但是会影响到线程并发执行的正确性。**

要想并发程序能够正确执行，必须保证原子性，有序性和可见性。一个不成立都不行。 


##### 3.java的内存模型
Java虚拟机规范中试图定义一种**Java内存模型（Java Memory Model，JMM）**来屏蔽各个硬件平台和操作系统的内存访问差异，以实现让Java程序在各种平台下都能达到一致的内存访问效果  

**在java内存模型中，也会存在缓存一致性问题和指令重排序的问题。**

Java内存模型规定所有的变量都是存在主存当中（类似于前面说的物理内存），**每个线程都有自己的工作内存（类似于前面的高速缓存）。线程对变量的所有操作都必须在工作内存中进行，而不能直接对主存进行操作。**并且每个线程不能访问其他线程的工作内存
在java内存模型中的原子性，可见性，有序性  

1. 原子性  
Java内存模型只保证了**基本读取和赋值**是原子性操作，如果要实现更大范围操作的原子性，可以通过synchronized和Lock来实现。由于synchronized和Lock能够保证任一时刻只有一个线程执行该代码块，那么自然就不存在原子性问题了，从而保证了原子性
2. 可见性  
对于可见性，**Java提供了volatile关键字来保证可见性。**
**当一个共享变量被volatile修饰时，它会保证修改的值会立即被更新到主存，当有其他线程需要读取时，它会去内存中读取新值。**  
普通的共享变量不保证可见性，因为不知道什么时候将更新过的值从工作内存写入主内存（物理内存中）  
3. 有序性  
**在java中，使用关键字volatile可以保证部分有序性**
通过synchronized和Lock来保证有序性，很显然，synchronized和Lock保证每个时刻是有一个线程执行同步代码，相当于是让线程顺序执行同步代码，自然就保证了有序性。


此外，**Java内存模型具备一些先天的“有序性”，即不需要通过任何手段就能够得到保证的有序性，这个通常也称为 happens-before 原则**。如果两个操作的执行次序无法从happens-before原则推导出来，那么它们就不能保证它们的有序性，虚拟机可以随意地对它们进行重排序。


**下面就来具体介绍下四条主要的happens-before原则（先行发生原则）：**

>程序次序规则：一个线程内，按照代码顺序，书写在前面的操作先行发生于书写在后面的操作 

            一段程序代码的执行在单个线程中看起来是有序的(可能执行的语句顺序发生了变换(重排序))，  
            但是最终结果正确.也就是保证单个线程的运行结果正确，但是多线程无法保证。  


>锁定规则：一个unLock操作先行发生于后面对同一个锁的lock操作  


            也就是说无论在单线程中还是多线程中，同一个锁如果出于被锁定的状态，那么必须先对锁进行了
            释放操作，后面才能继续进行lock操作。  
            
            
> **volatile变量规则：对一个变量的写操作先行发生于后面对这个变量的读操作**  


>传递规则：如果操作A先行发生于操作B，而操作B又先行发生于操作C，则可以得出操作A先行发生于操作C  


##### 4. Volatile关键字  
一旦一个共享变量(实例变量，静态变量)被 volatile关键字修饰之后  
1）保证了不同线程对这个变量进行操作时的可见性，即一个线程修改了某个变量的值，这新值对其他线程来说是立即可见的。  
2）禁止进行指令重排序。

**volatile关键字可以保证变量的可见性，但是无发保证堆变量操作的原子性(只能保证对变量赋值的原子性)  
对于有序性，volatile 关键字禁止指令重排序，可以保证一定的有序性**

　　volatile关键字禁止指令重排序有两层意思：

>当程序执行到volatile变量的读操作或者写操作时，在其前面的操作的更改肯定全部已经进行，且结果已经对后面的操作可见；在其后面的操作肯定还没有进行；


>在进行指令优化时，不能将在对volatile变量访问的语句放在其后面执行，也不能把volatile变量后面的语句放到其前面执行。


**下面展示一个使用volatile的例子，构建一个线程安全的单例模式**

```java
//使用volatile实现单例模式
class singleton{
    private  static singleton instance = null;
    private singleton(){
        System.out.println("创建了一个单例对象");
    }
    
    public static singleton getInstance(){ //没有考虑线程安全。因为可能多个线程同时执行语句1，同一个线程中语句1和语句2的执行顺序不对，可能导致多次创建对象
        if(instance == null){// 语句1
            instance = new singleton(); //语句2
        }
        return instance;
    }
}

Runnable r = ()->{
    try{
        Thread.sleep((int)(10*Math.random()));
    }catch(InterruptedException e){
        e.printStackTrace();
    }
    singleton s = singleton.getInstance();
    System.out.println("当前线程:"+Thread.currentThread().getId()+"获取的单例对象为:"+s);
};
for(int i=0;i<10;i++){
    Thread t = new Thread(r);
    t.start();
}
Thread.sleep(1000);
```

##### 14.5.10 原子性

在java.util.concurrent.atomic包中，**有许多类使用了高效的机器级的指令来保证其他操作的原子性**  
比如 AtomicInteger类提供的方法 incrementAndGet()方法和decrementAndGet()方法以原子方式提供了自增1和自减1  

```java
import java.util.concurrent.atomic.AtomicInteger;

AtomicInteger nextNumber = new AtomicInteger(0);
int x = 0;

Runnable r = ()->{
    int oldValue,newValue;
    x++;
    try{
        Thread.sleep((int)(10*Math.random()));
    }catch(InterruptedException e){
        e.printStackTrace();
    }
    do{
        oldValue = nextNumber.get();
        newValue = oldValue+1;
    }
    while(!nextNumber.compareAndSet(oldValue,newValue));

};

for(int i=0;i<10;i++){
    Thread t = new Thread(r);
    t.start();
}
Thread.sleep(1000);
System.out.println(x);
System.out.println(nextNumber.get());

```

##### 14.5.12 线程局部变量 ThreadLcoal


在并发编程的时候，成员变量如果不做任何处理其实是线程不安全的，各个线程都在操作同一个变量，显然是不行的，并且我们也知道volatile这个关键字也是不能保证线程安全的。那么在有一种情况之下，我们需要满足这样一个条件：**变量是同一个，但是每个线程都使用同一个初始值，也就是使用同一个变量的一个新的副本。这种情况之下ThreadLocal就非常使用**  

多线程访问同一个共享变量的时候容易出现并发问题，特别是多个线程对一个变量进行写入的时候，为了保证线程安全，一般使用者在访问共享变量的时候需要进行额外的同步措施才能保证线程安全性。**ThreadLocal是除了加锁这种同步方式之外的一种保证一种规避多线程访问出现线程不安全的方法**



**实现原理**  
实际上在每个Thread（线程)类中，都有一个ThreadLocal类内部的ThreadLocalMap类型的变量threadLocals, 

大致意思就是ThreadLocal提供了线程内存储变量的能力，这些变量不同之处在于每一个线程读取的变量是对应的互相独立的。通过get和set方法就可以得到当前线程对应的值。

做个不恰当的比喻，**从表面上看ThreadLocal相当于维护了一个map，key就是当前的线程，value就是需要存储的对象。**

这里的这个比喻是不恰当的，实际上是ThreadLocal的静态内部类ThreadLocalMap为每个Thread都维护了一个数组table，ThreadLocal确定了一个数组下标，而这个下标就是value存储的对应位置。。


```java
import java.lang.ThreadLocal;

ThreadLocal<Integer> count = new ThreadLocal();

Runnable r = ()->{
    count.set((int)(10*Math.random()));
    int oldValue;
    for(int i=0;i<1;i++){
        oldValue = count.get();
        System.out.println("当前线程为:"+Thread.currentThread().getId()+" 对应的count值为:"+oldValue);
        count.set(oldValue+1);
    }
};

for(int i =0;i<10;i++){
    Thread t = new Thread(r);
    t.start();
}
Thread.sleep(100);
```

##### 14.5.14 读/写锁  ReentrantReadWriteLock类


如果多线程从一个数据结构中读取数据的操作很频繁，但是写操作很少。那么使用读写锁时很方便的。  
它**允许多个线程对读操作共享访问，但是写操作必须是互斥访问的。**
而传统的synchronized关键字和ReentrantLock锁是读操作之间也是互斥的。  
同时 ReentrantReadWriteLock也是可重入锁，只要线程获得了读锁或者写锁，之后该线程可以一直获取该锁。


读写锁ReentrantReadWriteLock维护了一对锁，一个读锁readLock()，一个写锁writeLock()。

**读写锁的特性**
1. Java并发库中ReetrantReadWriteLock实现了ReadWriteLock接口并添加了可重入的特性
2. ReetrantReadWriteLock读写锁的效率明显高于synchronized关键字
3. ReetrantReadWriteLock读写锁的实现中，读锁使用共享模式；写锁使用独占模式，换句话说，读锁可以在没有写锁的时候被多个线程同时持有，写锁是独占的
4. ReetrantReadWriteLock读写锁的实现中，需要注意的，当有读锁时，写锁就不能获得；而当有写锁时，除了获得写锁的这个线程可以获得读锁外，其他线程不能获得读锁


**下面是读写锁的一个应用**

```java
import java.util.concurrent.locks.ReentrantReadWriteLock;
import java.util.concurrent.locks.Lock;
class CountTest{
    private int count;
    private  ReentrantReadWriteLock lock = new ReentrantReadWriteLock();
    private  Lock r = lock.readLock();
    private  Lock w = lock.writeLock();
    public CountTest(int i){
         count = i;
     }
     
    public int getCount(){
        r.lock();//获取读锁
        try{
            System.out.println("线程:"+Thread.currentThread().getId()+"开始读取数据");
            Thread.sleep((int)(10*Math.random()));
            return count;
            
        }catch(InterruptedException e){
            e.printStackTrace();
           
        }finally{
            System.out.println("线程:"+Thread.currentThread().getId()+"读取到的数据为:"+count);
            r.unlock();//释放读锁
        }
        return 0;
    }
    
    public void addOne(){
        w.lock();
        try{
            System.out.println("线程:"+Thread.currentThread().getId()+"正在写数据");
            Thread.sleep((int)(10*Math.random()));
            count++;
            System.out.println("线程:"+Thread.currentThread().getId()+"数据写入完成 "+"写入的数据为:"+count);
        }catch(InterruptedException e){
            e.printStackTrace();
        }finally{
            w.unlock();//释放写锁
        }
    }
 }

CountTest c = new CountTest(1);
Runnable r1 = ()->{
    c.getCount();
};
Runnable r2 = ()->{
    c.addOne();
};

for(int i=0;i<20;i++){
    if(i%5==0){//5的倍数，启动写线程
        Thread t = new Thread(r2);
        t.start();
    }else{//否则启动读线程
        Thread t = new Thread(r1);
        t.start();
    }
}
Thread.sleep(500);
```

### 14.8 Callabel和Future

Runnable对象相当于一个封装好的异步运行的任务(**异步指提交任务后，主线程继续执行，而任务在另一个线程中运行，直到执行完成后，再通知主线程任务执行完毕**)  

Runnable类似一个**没有参数和返回值的异步方法**  
而Callable<T> 接口类似，但是**他有返回值**。它只有一个方法call()。**call方法最终会返回一个类型为T的返回值。**
    
Future保存异步计算的结果。当启动一个异步任务后，将Future交给执行该任务的线程，然后主线程可以继续执行其他任务，当道任务完成后。利用Future对象就可以得到对应的计算结果。  
这意味着，**利用Future可以获得Runnable对象的最终计算结果**(本来Runnable是没有返回值得，也就无法在主线程中获得最终的结果，但是使用Future的get()方法，可以获取Runnable的最终计算结果）

FutureTask包装器就是这样的一种机制。它可以将Callable转换为Runnable和Future,FutureTask同时实现这两个接口。  


```java
import java.util.concurrent.Future;
import java.util.concurrent.Callable;
import java.util.concurrent.FutureTask;
int count = 0;
Callable<Integer> c = ()->{
    try{
        System.out.println("线程:"+Thread.currentThread().getId()+"开始执行Callable.");
        Thread.sleep(10);
    }catch(InterruptedException e){
        e.printStackTrace();
    }
    count++;
    System.out.println("线程:"+Thread.currentThread().getId()+"执行完成Callable.");
    return count;
};

FutureTask ft = new FutureTask(c); //FutureTask对象同时实现了Runnable接口和Future接口
Thread t = new Thread(ft);//将ft添加到创建的线程中
t.start();//开始执行子线程运行任务ft
//while(!ft.isDone());
System.out.println(ft.get());//这里如果任务没有完成，对get的调用会发生阻塞，知道任务完成。返回计算结果
```

### 14.9线程池

**java线程池是应用场景最多的并发框架**。几乎所有需要异步或者并发执行任务的程序都可以使用线程池。  
线程池的三个好处: （总的来说，主要是减少了线程的重复创建和销毁）   
> 降低资源消耗。重复利用已经创建的线程，从而降低线程创建和销毁的消耗。  

>提高响应的速度。当有任务到达时，不用再等待线程的创建就能够使用  

>提高线程的可管理性。使用线程池可以对线程进行统一的分配 调优和监控。不再需要无限制的去创建线程。

**线程池API的一部分**  


**线程池的应用场景**：  
如果应用需要创建大量的生命周期很短的线程，可以考虑使用线程池(Thread pool).  
一个线程池包含很多个空闲的线程，将Runnable对象提交给线程池，会分配一个空闲线程来调用run()方法来运行，运行
接收后run()方法退出但是线程不会死亡，而是变空闲，准备为下一个请求提供服务。  


java.util.concurrent.Executors类提供了多种**构建线程池的静态工厂方法**。常见的有5种:  

1. newSingleThreadExecutor: 创建一个单线程的线程池。**这个线程池只有一个线程在工作，也就是相当于单线程串行执行所有任务。**如果这个唯一的线程因为异常结束，那么会有一个新的线程来替代它。**此线程池保证所有任务的执行顺序按照任务的提交顺序执行。**

2. newFixedThreadPool: **创建固定大小(线程数量一定)的线程池**。每次提交一个任务就创建一个线程，直到线程达到线程池的最大大小。线程池的大小一旦达到最大值就会保持不变，如果某个线程因为执行异常而结束，那么线程池会补充一个新线程，空闲的线程会被保留。 **如果提交的任务多于线程数，就先将任务添加到队列中，等待空闲下来的线程。 

3. newCachedThreadPool: 创建一个可缓存的线程池(线程数量不固定)。如果线程池的大小超过了处理任务所需要的线程，那么就会回收部分空闲（60秒不执行任务）的线程，当任务数增加时，此线程池又可以智能的添加新线程来处理任务。**此线程池不会对线程池大小做限制，线程池大小完全依赖于操作系统（或者说JVM）能够创建的最大线程大小**。也就是当任务数小于现存的线程数时，如果存在超过60s没有执行的线程，那么这个线程可能会被回收。如果任务数大于现存的线程数，有会自动的创建新线程添加到线程池中。

4. newScheduledThreadPool: 可以用来代替java.util.Timer。  创建一个大小无限的线程池。**此线程池支持定时以及周期性执行任务的需求**

5. newSingleThreadScheduledExecutor: 用于执行定时或者周期性任务的单线程池。

#### 14.9.1 线程池

静态工厂方法 newCachedThreadPool(), newSingleThreadExecutor(), newFixedThreadPool()他们会**返回一个实现了ExecutorService接口的ThreadPoolExecutor类的对象。**

**一. 任务提交**  


使用submit()方法可以将Runnable对象或者Callable对象提交给ExecutorServer:(或者提交给他的实现类对象)  
使用**submit()方法提交后，会返回一个Future对象，可以查询任务的状态。**线程池会尽早的执行这个任务。  

如果提交的是Runnable对象，Future对象在任务执行完成后调用get方法只会返回null。  
如果提交的是Callable对象，Future对象在任务执行完成后调用get方法返回最终的计算结果。  

**二. Executor的生命周期**  
线程池Executor是**异步的执行任务**，因此任何时刻不能够直接获取提交的任务的状态。这些任务有可能已经完成，也有可能正在执行或者还在排队等待执行  
当一个线程池用完时，关闭线程池可能会出现两种情况:  
> **平缓关闭shutdown**  
    已经启动的任务全部执行完成，并且不再接收新的任务
   
> **立即关闭shutdownNow()**  
    立即结束已经启动的任务和正在等待的任务
    
**三. 线程池的状态**(包括四种状态，初始状态(new之前),运行状态(RUNNING),SHUTDOWN状态，STOP状态)  
1. 在构建完线程池之后，线程池就进入了运行状态RUNNING,随时接收任务准备执行  
2. 在线程池调用了shutdown()来平缓关闭时，线程池不再接收新的任务，并等待现存的任务完成执行。这一个过程为SHTUDOWN状态  
3. 如果调用shutdownNow()立即关闭线程池，线程池会进入STOP状态  
4. 一旦shutdown和shutdownNow方法执行完毕后，线程会进入TERMINAED状态

**四. 线程池的任务拒绝策略**  
为什么线程池会拒绝任务的提交:   
线程池有一个任务队列，用于缓存所有待处理的任务，正在处理的任务将从任务队列中移除。因此在任务队列长度有限的情况下就会出现新任务的拒绝处理问题，需要有一种策略来处理应该加入任务队列却因为队列已满无法加入的情况。另外在线程池关闭的时候也需要对任务加入队列操作进行额外的协调处理。  
**四种任务拒绝策略**(这四种策略是独立无关的)  
![img](http://www.blogjava.net/images/blogjava_net/xylz/Windows-Live-Writer/fe23666dd961_1435C/RejectedExecutionHandler_thumb.png)  


>直接丢弃任务  
    
        这个有两种方式，一种是丢弃当前将要添加到任务队列中的任务(DisCardPolicy),  
        另一种是抛弃队列中最久的任务(DiscardOldestPolicy)

>抛出异常方式（AbortPolicy)  
        
        当任务队列不再加入新任务时，会抛出一个运行时异常，中断调用者的处理过程  

>不进入线程池运行(CallerRunsPolicy)  
        
        这种方式任务将由调用者线程去执行  

##### 使用线程池的步骤
1. 调用Executors类中的静态工厂方法来获取线程池对象  
2. 调用线程池对象的submit方法来提交Runnable和Callable任务对象，并保存返回的Future对象  
3. 当不再提交任务时，调用shutdown方法停止线程池。

```java
import java.util.concurrent.Executors;
import java.util.concurrent.ThreadPoolExecutor;
import java.util.concurrent.ArrayBlockingQueue;
import java.util.concurrent.TimeUnit;
import java.util.concurrent.ThreadPoolExecutor.DiscardPolicy;
//创建线程池
ThreadPoolExecutor threadPool = new ThreadPoolExecutor(
1,1,0,TimeUnit.SECONDS, new ArrayBlockingQueue<Runnable>(1));
//设置线程池的任务拒绝策略为拒绝提交(如果队列满了，新提交的任务会被丢弃)
//threadPool.setRejectedExecutionHandler(new ThreadPoolExecutor.DiscardPolicy());

//设置线程池的任务拒绝策略为拒绝提交(如果队列满了，队列中等待最久的任务会被丢弃)
//threadPool.setRejectedExecutionHandler(new ThreadPoolExecutor.DiscardOldestPolicy());

//设置线程池的任务拒绝策略为调用线程(如果队列满了，新提交的任务会由调用者线程执行，而不在是线程池中的线程执行了)
threadPool.setRejectedExecutionHandler(new ThreadPoolExecutor.CallerRunsPolicy());

for(int i=0;i<10;i++){
    final int task_index = i;
    threadPool.submit(new Runnable(){
        public void run(){
            System.out.println("run task:"+task_index+" "+Thread.currentThread().getName());
            try{
                Thread.sleep(10);
            }catch(InterruptedException e){
                e.printStackTrace();
            }
            System.out.println("run over:"+task_index+" "+Thread.currentThread().getName());
        }
    });//向线程池提交任务
}
Thread.sleep(1000);
System.out.println("before shutdown");
threadPool.shutdown();
System.out.println("after shutdon，threadpool.isTerminated="+threadPool.isTerminated());
threadPool.awaitTermination(10L,TimeUnit.SECONDS);
System.out.println("after shutdon，threadpool.isTerminated="+threadPool.isTerminated());
```

### 周期性任务调度
java.util.Timer/TimerTask是唯一的内置任务调度方法，而且在很长一段时间里很热衷于使用这种方式进行**周期性任务调度**  

TimerTask实现了Runnable接口，它相当于是一个任务实例。  
 Timer和TimerTask可以一同工作。**Timer类用于安排任务，被安排的任务必须是TimerTask类的实例**  
 因此，为了安排任务，首先必须创建TimerTask对象，然后使用Timer实例安排任务的执行。  
 
创建定时任务最简单的方法就是扩展TimerTask类并重写run()方法。  
创建任务后，可以使用Timer类的对象安排任务的执行。具体的执行方法有多种，主要看调用的schedule()方法。  
比如延时执行，在特定的时间点开始执行，之后以特定的周期开始执行。



###### Timer/TimerTask的特性
Timer对任务的调度是基于绝对时间(**当前时间或者特定的一个时间点**)的。  
所有的TimerTask**只有一个线程TimerThread**来执行，因此同一时刻只有一个TimerTask在执行。  
任何一个TimerTask的执行异常都会导致Timer终止所有任务。  
由于**基于绝对时间并且是单线程执行**，因此在多个任务调度时，长时间执行的任务被执行后有可能导致短时间任务快速在短时间内被执行多次或者干脆丢弃多个任务。  


```java
import java.util.Timer;
import java.util.TimerTask;

public void log(String s){
    System.out.println(System.currentTimeMillis()+":"+s);
}
class myTimerTask extends TimerTask{
    private String name;
    public myTimerTask(String name){
        super();
        this.name = name;
    }
    public void run(){
        log("当前线程:"+Thread.currentThread().getId()+" 正在执行周期任务:"+name);
    }
}

myTimerTask r1 = new myTimerTask("r1"); //创建一个执行的任务
myTimerTask r2 = new myTimerTask("r2");
Timer t = new Timer(); //创建一个Timer实例
t.scheduleAtFixedRate(r1,1000L,200L); //传入待执行的任务，并设定执行的方案
t.schedule(r2,500L,200L);
log("开始执行周期任务");
Thread.sleep(1000);
```

#### 周期性任务调度
为了能改进Timer/TimerTask周期任务的不足，引入了**预定执行方法(周期性任务调度)** （ScheduledExecutorService接口)  


预定执行(延时一定的时间执行或者在特定的时间点开始执行),重复执行任务(以一定的周期重复执行任务)    
ScheduledExecutorService接口 允许使用**线程池机制**的java.util.Timer的泛化。

ScheduledExecutorService接口继承了ExecutorService接口，是一个完整的线程池对象。  
同时在ScheduledExecutorService接口中声明了四个用于执行周期性任务的schedule()方法  


>schedule(Runnable command,long delay, TimeUnit unit)  
        
            在指定的延迟时间一次性启动任务（Runnable），没有返回值。

>schedule(Callable<V> callable, long delay, TimeUnit unit)  
        
            在指定的延迟时间一次性启动任务（Callable）携带一个结果。
    
> scheduleAtFixedRate(Runnable command,long initialDelay,long period,TimeUnit unit)  
            
            创建并执行一个在给定初始延迟后首次启用的定期操作，后续操作具有给定的周期；也就是将在  
            initialDelay 后开始执行，然后在 initialDelay+period 后执行，接着在 initialDelay + 2 * 
            period 后执行，依此类推。如果任务的任何一个执行遇到异常，则后续执行都会被取消。否则，只能
            通过执行程序的取消或终止方法来终止该任务。如果此任务的任何一个执行要花费比其周期更长的时
            间，则将推迟后续执行，但不会同时执行.
            
            也就是每次任务开始的时间间隔相同

> scheduleWithFixedDelay(Runnable command,long initialDelay,long delay,TimeUnit unit) 

        创建并执行一个在给定初始延迟后首次启用的定期操作，随后，在每一次执行终止和下一次执行开始之间都
        存在给定的延迟。如果任务的任一执行遇到异常，就会取消后续执行。否则，只能通过执行程序的取消或终
        止方法来终止该任务。
        
        每次任务结束与下次任务开始之间具有相等间隔


**周期性任务的特性:**
1. 基于相对时间而不是绝对时间。以任务加入线程池(任务队列)的时间为起始时间
2. 基于线程池，允许多个线程执行任务，  
3. 基于线程池的ScheduledExecutorService在其中一个任务发生异常时会退出执行线程，但同时会有新的线程补充进来进行执行。ScheduledExecutorService可以做到不丢失任务。

```java
import java.util.concurrent.Executors;
import java.util.concurrent.ScheduledExecutorService;
import java.util.concurrent.TimeUnit;

public void log(String s){
    System.out.println(System.currentTimeMillis()+":"+s);
}

ScheduledExecutorService sPool = Executors.newScheduledThreadPool(3);//创建一个周期任务线程池
Runnable r1 = ()->{
    log("线程"+Thread.currentThread().getName()+": r1任务开始");
    try{
        Thread.sleep(200);
    }catch(InterruptedException e){
        e.printStackTrace();
    }
    log("线程"+Thread.currentThread().getName()+": r1任务结束");
};
Runnable r2 = ()->{
    log("线程"+Thread.currentThread().getName()+": r2任务开始");
    try{
        Thread.sleep(200);  //如果采用AtFixedRate的方式运行，当任务运行时间超过定义的周期时间时，上一个任务执行完成后，立即执行下一个任务
    }catch(InterruptedException e){
        e.printStackTrace();
    }
    log("线程"+Thread.currentThread().getName()+": r2任务结束");
};

sPool.scheduleAtFixedRate(r1,1000L,100L,TimeUnit.MILLISECONDS);
sPool.scheduleWithFixedDelay(r2,1000L,200L,TimeUnit.MILLISECONDS);
log("开始");
Thread.sleep(2000);
log("结束");
```

#### Fork-join框架

对于一些任务，它可以很自然的分解为许多子任务（分治算法）。这时候可以使用Fork-join框架。这是一个执行并行任务的框架。Fork就是把一个大任务切分成若干子任务并行执行，Join就是把这些子任务的执行结果合并。


```java
import java.util.concurrent.RecursiveTask;
import java.util.concurrent.ForkJoinPool;
import java.util.concurrent.Future;

public class Fibonacci extends RecursiveTask<Integer>{
    private int n;
    public Fibonacci(int n){
        this.n = n;
    }
    protected Integer compute(){
        if(n<=2)
            return 1;
        else{
            Fibonacii f1 = new Fibonacii(n-1);
            f1.fork();
            Fibonacii f2 = new Fibonacii(n-2);
            f2.fork();
            return f1.join()+f2.join();
        }
    }
}

ForkJoinPool forkjoinpool = new ForkJoinPool();//创建一个执行forkjoin任务的线程池
Fibonacci f = new Fibonacci(100);
Future<Integer> res = forkjoinpool.submit(f);
res.get();

```
