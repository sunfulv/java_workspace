**创建线程的两种方法**

1. 使用Runnable接口，通过实现Runnable接口来构建任务类，然后创建一个任务对象加入到线程中  调用线程start()来启动线程  
2. 直接定义一个Thread类的子类来定义线程

使用currentThread()静态方法可以返回当前执行线程Thread对象  

```java
import java.lang.Thread;
class Task implements Runnable{
    public void run(){
        System.out.println("正在运行的线程名称为:"+Thread.currentThread().getName());
        System.out.println("线程："+Thread.currentThread().getName()+"是否被中断:"+Thread.currentThread().isInterrupted());
        int i =0;
        try{
            while(!Thread.currentThread().isInterrupted()&& i++<10){ //如果线程没有被中断
                System.out.println("线程:"+Thread.currentThread().getName()+"正在执行，没有被中断");
                System.out.println("线程:"+Thread.currentThread().getName()+"的状态:"+Thread.currentThread().getState());
            }
        }
//         catch(InterruptedException e){
//             e.printStackTrace();
//         }
        finally{
            System.out.println("线程被终止"); //如果线程被中断，会跳出try语句，然后必然执行finally
            return;
        }
    }
}
public class Main{
    public static void main() throws InterruptedException{
        Task r = new Task();
        Thread t = new Thread(r);
        t.setName("task");
        t.start();

        Thread.sleep(3);
        System.out.println("正在运行的线程名称为:"+Thread.currentThread().getName());
        t.interrupt();
       

    }
}
Main m = new Main();
m.main()
```

```java
import java.lang.Thread;
class TaskThread extends Thread{
    public void run(){
        System.out.println("这是task线程");
    }
}
public class Main{
    public static void main() throws InterruptedException{
        TaskThread t = new TaskThread();
        t.start();
        Thread.sleep(500);
        System.out.println("这是主线程");
    }
}
Main m = new Main();
m.main()
```

### 14.4 线程属性

#### 1.线程优先级
在java中，每个线程都有一个优先级属性**priority**. 默认情况下，子线程的优先级继承父线程的优先级
priority属性的类型为int,默认的最低优先级为1最高优先级为10.主线程(main线程)优先级一般为5  
可以使用**setPriority方法**提高或者降低线程的优先级  
不过java虚拟机中的线程优先级依赖于系统，不同的系统映射的优先级方式不同。

```java
Task r = new Task();
Thread t = new Thread(r);
t.setPriority(3);
t.start();
System.out.println("主线程:"+Thread.currentThread().getName()+"的优先级为:"+Thread.currentThread().getPriority());
t.getPriority();
```

#### 2.守护线程(daemon thread)

在Java中有两类线程：User Thread(用户线程)、Daemon Thread(守护线程) 

用个比较通俗的比如，任何一个守护线程都是整个JVM中所有非守护线程的保姆：

**只要当前JVM实例中尚存在任何一个非守护线程没有结束，守护线程就全部工作；只有当最后一个非守护线程结束时，守护线程随着JVM一同结束工作。**  
Daemon的作用是为其他线程的运行提供便利服务，守护线程最典型的应用就是 GC (垃圾回收器)，它就是一个很称职的守护者。

User和Daemon两者几乎没有区别，唯一的不同之处就在于虚拟机的离开：如果 User Thread已经全部退出运行了，只剩下Daemon Thread存在了，虚拟机也就退出了。 因为没有了被守护者，Daemon也就没有工作可做了，也就没有继续运行程序的必要了。

```java
//在这个例子中，设置一个守护线程t,但是守护线程要延时100ms在执行，在这100ms中，主线程已经执行完退出了，
//这时由于没有其他的用户线程了，虚拟机直接退出，所以守护线程还没执行完成就终止了 

class TestRunnable implements Runnable{
    public void run() {
        try{
            Thread.sleep(100);//先休眠1ms,在输出
            System.out.println("当前执行的线程为:"+Thread.currentThread().getName());
        }catch(InterruptedException e){
            e.printStackTrace();
        }
    }
}

TestRunnable r = new TestRunnable();
Thread t = new Thread(r);
t.setDaemon(true);//将线程t设置为守护线程
t.start();
System.out.println("主线程的名字为:"+Thread.currentThread().getName());
```

#### 3.(未捕获异常)处理器

多线程运行不能按照顺序执行过程中捕获异常的方式来处理异常，异常会被直接抛出到控制台（由于线程的本质，使得你不能捕获从线程中逃逸的异常。一旦异常逃逸出任务的run方法，它就会向外传播到控制台，除非你采用特殊的形式捕获这种异常。），这样会让你很头疼，**无法捕捉到异常就无法处理异常而引发的问题**


在java中要捕捉多线程产生的异常，需要自定义异常处理器，并添加到线程中  

```java
class MyUncaughtExceptionHandler implements Thread.UncaughtExceptionHandler{
    @Override
    public void uncaughtException(Thread t,Throwable e){
        System.out.println("捕捉到了线程"+t.getName()+"抛出的异常"+e);
    }
}
class TestException implements Runnable{
    public void run(){
        //System.out.println("当前线程为:"+Thread.currentThread().getName());
        throw new RuntimeException(); //这里子线程抛出了异常，但是在主线程中并没有捕获到
    }
}
//没有设置捕获异常处理器
/*try{
    TestException r = new TestException();
    Thread t = new Thread(r);
    t.start();
}catch(RuntimeException e){
    System.out.print("捕获到了异常");//子线程抛出异常，但是主线程无法捕捉到这个异常
}*/
//设置了捕获异常处理器
TestException r = new TestException();
Thread t = new Thread(r);
t.setUncaughtExceptionHandler(new MyUncaughtExceptionHandler());//为线程设置一个未捕获异常处理器，来捕捉线程抛出的受查异常
t.start();
```

#### 14.5同步与锁

##### 1.重入锁 ReentrantLock
重入锁的意思是这个锁能够支持一个线程对资源重复加锁。（重入锁还支持获取锁时的公平和非公平性选择）  
也就是在调用**锁的lock()方法**时，对于已经获取到该锁的线程 ，能够再次调用lock()方法而不被阻塞  
**也就是说线程可以重复获得已经持有的锁**，锁保持一个持有计数(hold count)来跟踪方法的嵌套调用

```java
import java.util.concurrent.locks.ReentrantLock;
class Count {
    int count;
    ReentrantLock myLock = new ReentrantLock();
    public Count(int i){
        count = i;
    }
    public int countAddOne(){//写入数据,并返回新的值
        //myLock.lock();
        try{
                count++;
                return count;
        }finally{
          //      myLock.unlock();
        }
    }
}

class Main{
    public static void main(){
        Count c = new Count(10);
        Runnable r = ()->{  try{
                            int i = 0;
                          while(i++<2){
                              int currentCount = c.countAddOne();

                              System.out.println("当前线程为:"+Thread.currentThread().getName()+"线程状态为:"+Thread.currentThread().getState()+"当前计数器c的值为:"+currentCount);
                              Thread.sleep(5);
                            }}catch(InterruptedException e){
                                e.printStackTrace();
                            }
                        };

        for(int i=0;i<10;i++){
            Thread t = new Thread(r);
            t.start();
        }
    }
}
Main.main();
Thread.sleep(1000);
```

#### 14.5.4 条件对象（条件变量）
线程除了会共同竞争某一个资源的使用权(使用锁机制来保证互斥资源同一时刻只有一个线程使用)  
有时还需要**相互合作**来完成特定的任务(线程按照一定的顺序执行)，**某一个线程在另一个线程执行完成后，才能继续执行**

java中，使用等待/通知机制来实现这个功能  


**java的Condition对象**  
Condition用于管理已经获得了一个锁但是却不能做有用工作的线程。由Lock对象的newCondition()方法一个与当前Lock对象相关联的条件对象，共同管理该锁相关的对象和线程。

当线程已经获得锁，但是因为条件限制，无法继续工作的时候，调用Condition.await()方法，此线程进入该条件对象的等待集。当锁可用时，该线程不能马上解除阻塞。相反，它处于阻塞状态，直到另个线程调用同一条件上的signalAll方法时为止，这个方法可以激活所有等待的线程。Condition还有一个Signal方法，该方法是随机激活一个线程。
当一个线程调用awiat()方法时，它无法重新激活自身。 只能寄希望给予其他的线程。如果没有其他线程激活，那他就永远无法在运行了，这就是死锁现象。
SignalAll不会立即激活一个等待线程，它仅仅解除等待线程的阻塞，以便这些线程可以在当前线程退出同步方法之后，通过竞争实现对对象的访问

```java
//实现一个有界队列，如果队列为空，线程获取元素操作会导致线程堵塞，直到队列不空。
//如果队列满，线程插入元素会导致线程阻塞，知道队列中出现空位。
import java.util.concurrent.locks.ReentrantLock;
import java.util.concurrent.locks.Condition;
public class BoundedQueue<T>{
    Object[] queue;
    ReentrantLock myLock = new ReentrantLock();
    Condition emptyCondition = myLock.newCondition();
    Condition fullCondition = myLock.newCondition();
    int size;
    int front;
    int rear;
    public BoundedQueue(int length) {
        queue = new Object[length+1];//预留出一个空位置
        size = 0;
    }
    
    public boolean isEmpty(){
        if(size == 0) return true;
        return false;
    }
    public void add(T ele){
        myLock.lock();//先获取锁
        try{
                while(size == queue.length-1)//循环检测队列是否满
                    emptyCondition.await();//等待队列出现空闲
                queue[rear] = ele;
                size++;
                rear = (rear+1)%queue.length;
                System.out.println("当前线程id为:"+Thread.currentThread().getId()+" 当前插入的元素为:"+ele);
                System.out.println(Arrays.toString(queue));
                fullCondition.signalAll();//队列中出现了元素，唤醒那些因为队列没有元素而阻塞的线程
            }catch(InterruptedException e){
                e.printStackTrace();
            }finally{
                myLock.unlock();//操作完成释放锁
            }
    }
    @SuppressWarnings("unchecked")
    public T delete(){
        myLock.lock();
        T oldValue = null;
        try{
                while(size == 0)//循环检测队列是否为空,这是为了防止过早或者意外的通知
                    fullCondition.await();
                oldValue = (T)queue[front];
                front = (front+1)%queue.length;
                size--;
                System.out.println("当前线程id为:"+Thread.currentThread().getId()+" 删除的元素为:"+oldValue);
                emptyCondition.signalAll();
            }catch(InterruptedException e){
                e.printStackTrace();
            }finally{
                myLock.unlock();
            }
        return oldValue;
    }
}

BoundedQueue<Integer> myQueue = new BoundedQueue(3);
Runnable r1 =()->{
    for(int i=0;i<10;i++){
        myQueue.add(i*i);
        try{
            Thread.sleep(10);
        }catch(InterruptedException e){
            e.printStackTrace();
        }
    }
};
Runnable r2 = ()->{
    for(int i = 0;i<10;i++){
        myQueue.delete();
        try{
            Thread.sleep(10);
        }catch(InterruptedException e){
            e.printStackTrace();
        }
    }
};

Thread t1 = new Thread(r1);
Thread t2 = new Thread(r2);
Thread t3 = new Thread(r2);
System.out.printf("当前创建的子线程:%d.%d,%d\n",t1.getId(),t2.getId(),t3.getId());
t3.start();
t2.start();
t1.start();


Thread.sleep(1000);
```
