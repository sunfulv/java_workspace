#### 2.守护线程(daemon thread)

在Java中有两类线程：User Thread(用户线程)、Daemon Thread(守护线程) 

用个比较通俗的比如，任何一个守护线程都是整个JVM中所有非守护线程的保姆：

**只要当前JVM实例中尚存在任何一个非守护线程没有结束，守护线程就全部工作；只有当最后一个非守护线程结束时，守护线程随着JVM一同结束工作。**  
Daemon的作用是为其他线程的运行提供便利服务，守护线程最典型的应用就是 GC (垃圾回收器)，它就是一个很称职的守护者。

User和Daemon两者几乎没有区别，唯一的不同之处就在于虚拟机的离开：如果 User Thread已经全部退出运行了，只剩下Daemon Thread存在了，虚拟机也就退出了。 因为没有了被守护者，Daemon也就没有工作可做了，也就没有继续运行程序的必要了。

```java
//在这个例子中，设置一个守护线程t,但是守护线程要延时100ms在执行，在这100ms中，主线程已经执行完退出了，
//这时由于没有其他的用户线程了，虚拟机直接退出，所以守护线程还没执行完成就终止了 

class TestRunnable implements Runnable{
    public void run() {
        try{
            Thread.sleep(100);//先休眠1ms,在输出
            System.out.println("当前执行的线程为:"+Thread.currentThread().getName());
        }catch(InterruptedException e){
            e.printStackTrace();
        }
    }
}

TestRunnable r = new TestRunnable();
Thread t = new Thread(r);
t.setDaemon(true);//将线程t设置为守护线程
t.start();
System.out.println("主线程的名字为:"+Thread.currentThread().getName());
```

#### 14.5.4 条件对象（条件变量）
线程除了会共同竞争某一个资源的使用权(使用锁机制来保证互斥资源同一时刻只有一个线程使用)  
有时还需要**相互合作**来完成特定的任务(线程按照一定的顺序执行)，**某一个线程在另一个线程执行完成后，才能继续执行**

java中，使用等待/通知机制来实现这个功能  


**java的Condition对象**  
Condition用于管理已经获得了一个锁但是却不能做有用工作的线程。由Lock对象的newCondition()方法一个与当前Lock对象相关联的条件对象，共同管理该锁相关的对象和线程。

当线程已经获得锁，但是因为条件限制，无法继续工作的时候，调用Condition.await()方法，此线程进入该条件对象的等待集。当锁可用时，该线程不能马上解除阻塞。相反，它处于阻塞状态，直到另个线程调用同一条件上的signalAll方法时为止，这个方法可以激活所有等待的线程。Condition还有一个Signal方法，该方法是随机激活一个线程。
当一个线程调用awiat()方法时，它无法重新激活自身。 只能寄希望给予其他的线程。如果没有其他线程激活，那他就永远无法在运行了，这就是死锁现象。
SignalAll不会立即激活一个等待线程，它仅仅解除等待线程的阻塞，以便这些线程可以在当前线程退出同步方法之后，通过竞争实现对对象的访问

**Condition对象的await()方法和singal()/signalAll()方法**

Lock锁对象可以创建多个Condition条件对象。每个条件对象，才能调用signal()和await()方法  
当线程如果因为某个条件不满足(使用while循环来不断的检查是否满足条件).条件对象调用await()方法，然后线程进入这个条件的等待队列，然后释放锁。  当锁可用的时候，线程不能马上解除阻塞，需要另一个线程调用同一个条件对象上的signalAll()方法时，才能激活该线程，这是线程又变得可运行态，重新试图获得锁。获得锁止呕可以从await()返回，然后从组册的位置继续运行。

```java
//实现一个有界队列，如果队列为空，线程获取元素操作会导致线程堵塞，直到队列不空。
//如果队列满，线程插入元素会导致线程阻塞，知道队列中出现空位。
import java.util.concurrent.locks.ReentrantLock;
import java.util.concurrent.locks.Condition;
public class BoundedQueue<T>{
    Object[] queue;
    ReentrantLock myLock = new ReentrantLock();
    Condition emptyCondition = myLock.newCondition();
    Condition fullCondition = myLock.newCondition();
    int size;
    int front;
    int rear;
    public BoundedQueue(int length) {
        queue = new Object[length+1];//预留出一个空位置
        size = 0;
    }
    
    public boolean isEmpty(){
        if(size == 0) return true;
        return false;
    }
    public void add(T ele){
        myLock.lock();//先获取锁
        try{
                while(size == queue.length-1)//循环检测队列是否满
                    emptyCondition.await();//等待队列出现空闲
                queue[rear] = ele;
                size++;
                rear = (rear+1)%queue.length;
                System.out.println("当前线程id为:"+Thread.currentThread().getId()+" 当前插入的元素为:"+ele+" 当前队列元素:"+Arrays.toString(queue));
                
                fullCondition.signalAll();//队列中出现了元素，唤醒那些因为队列没有元素而阻塞的线程
            }catch(InterruptedException e){
                e.printStackTrace();
            }finally{
                myLock.unlock();//操作完成释放锁
            }
    }
    @SuppressWarnings("unchecked")
    public T delete(){
        myLock.lock();
        T oldValue = null;
        try{
                while(size == 0)//循环检测队列是否为空,这是为了防止过早或者意外的通知
                    fullCondition.await();
                oldValue = (T)queue[front];
                queue[front] = null;
                front = (front+1)%queue.length;
                size--;
                System.out.println("当前线程id为:"+Thread.currentThread().getId()+" 删除的元素为:"+oldValue+"当前队列元素为:"+Arrays.toString(queue));
                emptyCondition.signalAll();
            }catch(InterruptedException e){
                e.printStackTrace();
            }finally{
                myLock.unlock();
            }
        return oldValue;
    }
}

BoundedQueue<Integer> myQueue = new BoundedQueue(5);
Runnable r1 =()->{
    for(int i=0;i<10;i++){
        myQueue.add(i*i);
        try{
            Thread.sleep(10);
        }catch(InterruptedException e){
            e.printStackTrace();
        }
    }
};
Runnable r2 = ()->{
    for(int i = 0;i<10;i++){
        myQueue.delete();
        try{
            Thread.sleep(10);
        }catch(InterruptedException e){
            e.printStackTrace();
        }
    }
};

Thread t1 = new Thread(r1);
Thread t2 = new Thread(r2);
Thread t3 = new Thread(r2);
System.out.printf("当前创建的子线程:%d.%d,%d\n",t1.getId(),t2.getId(),t3.getId());
t3.start();
t2.start();
t1.start();

    
Thread.sleep(1000);
```

此外java内部提供了一种内部机制来实现同步。  
Java对象中每个对象都有一个内部锁，如果一个方法使用synchronized关键字声明，那么对象的锁将保护整个方法。**要调用这个方法，线程必须获得内部锁。**

```java
class Counter{
    int count;
    public Counter(int i){
        count = i;
    }
    public synchronized void addOne(){
            count++;
            System.out.println("当前线程id为:"+Thread.currentThread().getId()+" 执行加一操作 "+"当前计数器的值为:"+count);
            notify(); //计数器大于0，唤醒阻塞的队列
            //Thread.sleep(1);

    }
    public synchronized void minusOne(){
        try{
            while(count<=0)
                wait();//如果计数器小于0，等待
            count--;
            System.out.println("当前线程id为:"+Thread.currentThread().getId()+" 执行减一操作"+" 当前计数器的值为:"+count);
            //Thread.sleep(1);
        }catch(InterruptedException e){
            e.printStackTrace();
        }
    } 
}

Counter c = new Counter(0);
Runnable r1 = ()->{
    for(int i=0;i<100;i++){
        c.addOne();
    }
};
Runnable r2 = ()->{
    for(int j=0;j<100;j++){
        c.minusOne();
    }
};
Thread t1 = new Thread(r1);
Thread t2 = new Thread(r2);
t2.start();
t1.start();
Thread.sleep(1000);
```

#### 14.5.6 volatile 关键字

　　一旦一个共享变量（类的成员变量、类的静态成员变量）被volatile修饰之后，那么就具备了两层语义：

　　1）保证了不同线程对这个变量进行操作时的可见性，即一个线程修改了某个变量的值，这新值对其他线程来说是立即可见的。   
　　2）禁止进行指令重排序。
  
  
**volatile关键字是与Java的内存模型有关的，因此在讲述volatile关键之前，我们先来了解一下与内存模型相关的概念和知识**

##### 1.内存模型的相关概念
当程序在运行过程中，会将运算需要的数据**从主存复制一份到CPU的高速缓存**当中，那么CPU进行计算时就可以直接从它的高速缓存读取数据和向其中写入数据，当运算结束之后，再将高速缓存中的数据刷新到主存当中。

但是对于多核系统，每个CPU都有自己的高速缓存，如果变量在多个高速缓存中都有缓存(多线程编程中)，**会导致缓存不一致问题**  
为了解决缓存不一致问题，一种可行的办法是定义**缓存一致性协议。**  


缓存一致性协议。最出名的就是Intel 的MESI协议，MESI协议保证了每个缓存中使用的共享变量的副本是一致的。它核心的思想是：**当CPU写数据时，如果发现操作的变量是共享变量，即在其他CPU中也存在该变量的副本，会发出信号通知其他CPU将该变量的缓存行置为无效状态，因此当其他CPU需要读取这个变量时，发现自己缓存中缓存该变量的缓存行是无效的，那么它就会从内存重新读取**

##### 2.并发编程的三个问题

>原子性
>可见性
>有序性

1. 原子性
即一个操作或者多个操作 要么全部执行并且执行的过程不会被任何因素打断，要么就都不执行。  
2. 可见性
即一个操作或者多个操作 要么全部执行并且执行的过程不会被任何因素打断，要么就都不执行。
3. 有序性
有序性：即程序执行的顺序按照代码的先后顺序执行。  
**但是有时编译器会在不改变最终运算结果的情况下改变指令的执行顺序(也就是指令重排序)**  
但是如果是多线程程序，改变指令的执行顺序可能会导致最终结果发生改变，从而得不到正确结果。    
**指令重排序不会影响单个线程的执行，但是会影响到线程并发执行的正确性。**

要想并发程序能够正确执行，必须保证原子性，有序性和可见性。一个不成立都不行。 


##### 3.java的内存模型
Java虚拟机规范中试图定义一种**Java内存模型（Java Memory Model，JMM）**来屏蔽各个硬件平台和操作系统的内存访问差异，以实现让Java程序在各种平台下都能达到一致的内存访问效果  

**在java内存模型中，也会存在缓存一致性问题和指令重排序的问题。**

Java内存模型规定所有的变量都是存在主存当中（类似于前面说的物理内存），**每个线程都有自己的工作内存（类似于前面的高速缓存）。线程对变量的所有操作都必须在工作内存中进行，而不能直接对主存进行操作。**并且每个线程不能访问其他线程的工作内存
在java内存模型中的原子性，可见性，有序性  

1. 原子性  
Java内存模型只保证了**基本读取和赋值**是原子性操作，如果要实现更大范围操作的原子性，可以通过synchronized和Lock来实现。由于synchronized和Lock能够保证任一时刻只有一个线程执行该代码块，那么自然就不存在原子性问题了，从而保证了原子性
2. 可见性  
对于可见性，**Java提供了volatile关键字来保证可见性。**
**当一个共享变量被volatile修饰时，它会保证修改的值会立即被更新到主存，当有其他线程需要读取时，它会去内存中读取新值。**  
普通的共享变量不保证可见性，因为不知道什么时候将更新过的值从工作内存写入主内存（物理内存中）  
3. 有序性  
**在java中，使用关键字volatile可以保证部分有序性**
通过synchronized和Lock来保证有序性，很显然，synchronized和Lock保证每个时刻是有一个线程执行同步代码，相当于是让线程顺序执行同步代码，自然就保证了有序性。


此外，**Java内存模型具备一些先天的“有序性”，即不需要通过任何手段就能够得到保证的有序性，这个通常也称为 happens-before 原则**。如果两个操作的执行次序无法从happens-before原则推导出来，那么它们就不能保证它们的有序性，虚拟机可以随意地对它们进行重排序。


**下面就来具体介绍下四条主要的happens-before原则（先行发生原则）：**

>程序次序规则：一个线程内，按照代码顺序，书写在前面的操作先行发生于书写在后面的操作 

            一段程序代码的执行在单个线程中看起来是有序的(可能执行的语句顺序发生了变换(重排序))，  
            但是最终结果正确.也就是保证单个线程的运行结果正确，但是多线程无法保证。  


>锁定规则：一个unLock操作先行发生于后面对同一个锁的lock操作  


            也就是说无论在单线程中还是多线程中，同一个锁如果出于被锁定的状态，那么必须先对锁进行了
            释放操作，后面才能继续进行lock操作。  
            
            
> **volatile变量规则：对一个变量的写操作先行发生于后面对这个变量的读操作**  


>传递规则：如果操作A先行发生于操作B，而操作B又先行发生于操作C，则可以得出操作A先行发生于操作C  


##### 4. Volatile关键字  
一旦一个共享变量(实例变量，静态变量)被 volatile关键字修饰之后  
1）保证了不同线程对这个变量进行操作时的可见性，即一个线程修改了某个变量的值，这新值对其他线程来说是立即可见的。  
2）禁止进行指令重排序。

**volatile关键字可以保证变量的可见性，但是无发保证堆变量操作的原子性(只能保证对变量赋值的原子性)  
对于有序性，volatile 关键字禁止指令重排序，可以保证一定的有序性**

　　volatile关键字禁止指令重排序有两层意思：

>当程序执行到volatile变量的读操作或者写操作时，在其前面的操作的更改肯定全部已经进行，且结果已经对后面的操作可见；在其后面的操作肯定还没有进行；


>在进行指令优化时，不能将在对volatile变量访问的语句放在其后面执行，也不能把volatile变量后面的语句放到其前面执行。


**下面展示一个使用volatile的例子，构建一个线程安全的单例模式**

```java
//使用volatile实现单例模式
class singleton{
    private  static singleton instance = null;
    private singleton(){
        System.out.println("创建了一个单例对象");
    }
    
    public static singleton getInstance(){ //没有考虑线程安全。因为可能多个线程同时执行语句1，同一个线程中语句1和语句2的执行顺序不对，可能导致多次创建对象
        if(instance == null){// 语句1
            instance = new singleton(); //语句2
        }
        return instance;
    }
}

Runnable r = ()->{
    try{
        Thread.sleep((int)(10*Math.random()));
    }catch(InterruptedException e){
        e.printStackTrace();
    }
    singleton s = singleton.getInstance();
    System.out.println("当前线程:"+Thread.currentThread().getId()+"获取的单例对象为:"+s);
};
for(int i=0;i<10;i++){
    Thread t = new Thread(r);
    t.start();
}
Thread.sleep(1000);
```

##### 14.5.10 原子性

在java.util.concurrent.atomic包中，**有许多类使用了高效的机器级的指令来保证其他操作的原子性**  
比如 AtomicInteger类提供的方法 incrementAndGet()方法和decrementAndGet()方法以原子方式提供了自增1和自减1  

```java
import java.util.concurrent.atomic.AtomicInteger;

AtomicInteger nextNumber = new AtomicInteger(0);
int x = 0;

Runnable r = ()->{
    int oldValue,newValue;
    x++;
    try{
        Thread.sleep((int)(10*Math.random()));
    }catch(InterruptedException e){
        e.printStackTrace();
    }
    do{
        oldValue = nextNumber.get();
        newValue = oldValue+1;
    }
    while(!nextNumber.compareAndSet(oldValue,newValue));

};

for(int i=0;i<10;i++){
    Thread t = new Thread(r);
    t.start();
}
Thread.sleep(1000);
System.out.println(x);
System.out.println(nextNumber.get());

```

##### 14.5.12 线程局部变量 ThreadLcoal


在并发编程的时候，成员变量如果不做任何处理其实是线程不安全的，各个线程都在操作同一个变量，显然是不行的，并且我们也知道volatile这个关键字也是不能保证线程安全的。那么在有一种情况之下，我们需要满足这样一个条件：**变量是同一个，但是每个线程都使用同一个初始值，也就是使用同一个变量的一个新的副本。这种情况之下ThreadLocal就非常使用**  

多线程访问同一个共享变量的时候容易出现并发问题，特别是多个线程对一个变量进行写入的时候，为了保证线程安全，一般使用者在访问共享变量的时候需要进行额外的同步措施才能保证线程安全性。**ThreadLocal是除了加锁这种同步方式之外的一种保证一种规避多线程访问出现线程不安全的方法**



**实现原理**  
实际上在每个Thread（线程)类中，都有一个ThreadLocal类内部的ThreadLocalMap类型的变量threadLocals, 

大致意思就是ThreadLocal提供了线程内存储变量的能力，这些变量不同之处在于每一个线程读取的变量是对应的互相独立的。通过get和set方法就可以得到当前线程对应的值。

做个不恰当的比喻，**从表面上看ThreadLocal相当于维护了一个map，key就是当前的线程，value就是需要存储的对象。**

这里的这个比喻是不恰当的，实际上是ThreadLocal的静态内部类ThreadLocalMap为每个Thread都维护了一个数组table，ThreadLocal确定了一个数组下标，而这个下标就是value存储的对应位置。。


```java
import java.lang.ThreadLocal;

ThreadLocal<Integer> count = new ThreadLocal();

Runnable r = ()->{
    count.set((int)(10*Math.random()));
    int oldValue;
    for(int i=0;i<1;i++){
        oldValue = count.get();
        System.out.println("当前线程为:"+Thread.currentThread().getId()+" 对应的count值为:"+oldValue);
        count.set(oldValue+1);
    }
};

for(int i =0;i<10;i++){
    Thread t = new Thread(r);
    t.start();
}
Thread.sleep(100);
```

##### 14.5.14 读/写锁  ReentrantReadWriteLock类


如果多线程从一个数据结构中读取数据的操作很频繁，但是写操作很少。那么使用读写锁时很方便的。  
它**允许多个线程对读操作共享访问，但是写操作必须是互斥访问的。**
而传统的synchronized关键字和ReentrantLock锁是读操作之间也是互斥的。  
同时 ReentrantReadWriteLock也是可重入锁，只要线程获得了读锁或者写锁，之后该线程可以一直获取该锁。


读写锁ReentrantReadWriteLock维护了一对锁，一个读锁readLock()，一个写锁writeLock()。

**读写锁的特性**
1. Java并发库中ReetrantReadWriteLock实现了ReadWriteLock接口并添加了可重入的特性
2. ReetrantReadWriteLock读写锁的效率明显高于synchronized关键字
3. ReetrantReadWriteLock读写锁的实现中，读锁使用共享模式；写锁使用独占模式，换句话说，读锁可以在没有写锁的时候被多个线程同时持有，写锁是独占的
4. ReetrantReadWriteLock读写锁的实现中，需要注意的，当有读锁时，写锁就不能获得；而当有写锁时，除了获得写锁的这个线程可以获得读锁外，其他线程不能获得读锁


**下面是读写锁的一个应用**

```java
import java.util.concurrent.locks.ReentrantReadWriteLock;
import java.util.concurrent.locks.Lock;
class CountTest{
    private int count;
    private  ReentrantReadWriteLock lock = new ReentrantReadWriteLock();
    private  Lock r = lock.readLock();
    private  Lock w = lock.writeLock();
    public CountTest(int i){
         count = i;
     }
     
    public int getCount(){
        r.lock();//获取读锁
        try{
            System.out.println("线程:"+Thread.currentThread().getId()+"开始读取数据");
            Thread.sleep((int)(10*Math.random()));
            return count;
            
        }catch(InterruptedException e){
            e.printStackTrace();
           
        }finally{
            System.out.println("线程:"+Thread.currentThread().getId()+"读取到的数据为:"+count);
            r.unlock();//释放读锁
        }
        return 0;
    }
    
    public void addOne(){
        w.lock();
        try{
            System.out.println("线程:"+Thread.currentThread().getId()+"正在写数据");
            Thread.sleep((int)(10*Math.random()));
            count++;
            System.out.println("线程:"+Thread.currentThread().getId()+"数据写入完成 "+"写入的数据为:"+count);
        }catch(InterruptedException e){
            e.printStackTrace();
        }finally{
            w.unlock();//释放写锁
        }
    }
 }

CountTest c = new CountTest(1);
Runnable r1 = ()->{
    c.getCount();
};
Runnable r2 = ()->{
    c.addOne();
};

for(int i=0;i<20;i++){
    if(i%5==0){//5的倍数，启动写线程
        Thread t = new Thread(r2);
        t.start();
    }else{//否则启动读线程
        Thread t = new Thread(r1);
        t.start();
    }
}
Thread.sleep(500);
```
